```python
def get_range(mask):
    if mask[1] == 28:
        return 16
    elif mask[1]==29:
        return 8
    elif mask[1] ==30:
        return 4
    elif mask[1]== 32:
        #print("pass")
        return 1
    
```

```python
#ip1="192.168.1.0/30"
def getr_ips(net):
    mask=list(map(int,((net.split(".")[3]).split("/"))))
    #print(mask)
    r=get_range(mask)
    #print("we have  ============{}================ IPs in this network ".format(r))
    range_ip=[]
    if r:
        #print("subnetMask {}".format(net))
        ip=net.split(".")[0:4]
        ip[3]=ip[3].split("/")[0]
        range_ip.append(".".join(ip)+"/")
        for i in range(r-1):
            ip[3]=str(int(ip[3])+1)
            #print(".".join(ip))
            range_ip.append(".".join(ip)+"/")
    return range_ip

getr_ips("192.168.1.0/29")
```
